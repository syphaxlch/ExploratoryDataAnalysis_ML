# Chargement de dictionary.yml

```python
with open('data/dictionary.yml', 'r') as file:
    data_yaml = yaml.safe_load(file)
```

```python
df_dict = pd.DataFrame(data_yaml['data-schema'])
df_dict
```

```python
# Filtrer les √©l√©ments dont le type est un struct
df_struct = df_dict[df_dict['type'].str.contains('struct', na=False)]

# Afficher les √©l√©ments struct
print(df_struct)
```
